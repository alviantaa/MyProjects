## Identify the data types of each column


Explore the dataset and identify the data types of each column.


Print the datatype of all columns.

Print the mean age of the survey participants.


```python
# Check the bins of values for the Age column
df['Age'].unique()
```

```python
age_mapping = {
    'Under 18 years old': 17,   # Assume 17 for "Under 18"
    '18-24 years old': 21,      # Midpoint (18+24)/2
    '25-34 years old': 29.5,    # Midpoint (25+34)/2
    '35-44 years old': 39.5,    # Midpoint (35+44)/2
    '45-54 years old': 49.5,    # Midpoint (45+54)/2
    '55-64 years old': 59.5,    # Midpoint (55+64)/2
    '65 years or older': 70,    # Assume 70 for "65+"
    'Prefer not to say': np.nan # Exclude from calculation
}

# Add column for converted Age column to midpoints
df['Age Midpoint'] = df['Age'].map(age_mapping)
print(df['Age Midpoint'])
```

```python
mean_age = df['Age Midpoint'].mean()

print(f"Average Age: {mean_age:.2f}")
```

The dataset is the result of a world wide survey. Print how many unique countries are there in the Country column.


```python
print('Total Country', df['Country'].nunique())

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df['Country'].value_counts())
```
